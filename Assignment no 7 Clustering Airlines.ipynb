```python
airl=pd.read_excel(r'C:\Users\hindu\Downloads\EastWestAirlines (2).xlsx',sheet_name='data')
```

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)

# Normalized data frame (considering the numerical part of data)
df_norm = norm_func(airl.iloc[:,1:])


df_norm.head(10)  # Top 10 rows
```

```python
# Selecting 5 clusters from the above scree plot which is the optimum number of clusters 
model=KMeans(n_clusters=5) 
model.fit(df_norm)

model.labels_ # getting the labels of clusters assigned to each row
```

```python
md=pd.Series(model.labels_)  # converting numpy array into pandas series object 
airl['clust']=md # creating a  new column and assigning it to new column 
df_norm.head()
```

```python
airl.iloc[:,1:7].groupby(airl.clust).mean()
```

```python
airl.head()
```

```python
Wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(df_norm)
    Wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),Wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('Wcss')
plt.show()
```

# DBSCAN

```python
#Import the libraries
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import .csv file and convert it to a DataFrame object
airl=pd.read_excel(r'C:\Users\hindu\Downloads\EastWestAirlines (2).xlsx',sheet_name='data')

print(airl.head())
```

```python
print(airl.info())
```

```python
airl.drop(['ID#','Award?'],axis=1,inplace=True)
```

```python
airl
```

```python
array=airl.values
```

```python
array
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array)
```

```python

X
```

```python
dbscan = DBSCAN(eps=1, min_samples=12)
dbscan.fit(X)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
cl
```

```python
pd.concat([airl,cl],axis=1)
```
